```python
import tensorflow as tf
import numpy as np
import plotly.graph_objects as go
from tensorflow import keras
from dataloader import FuncLoader
```

## 数据集是sin的拟合，范围\[-10,10\]，测试模型是神经网络，训练：验证=8：2

```python
dl = FuncLoader(name='sin')
train_dataset,validation_dataset = dl.load_data()
```

### 0001:三层神经网络，最后激活为sigmoid，传统的dnn模型。
测试在\[-10，10\], loss:mse, optimzier:adam

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(16, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])
model.compile(optimizer='adam',
                loss='mse',
                metrics=['mae'])
model.fit(train_dataset, epochs=150,validation_data=validation_dataset)
test_x = np.linspace(-10,10,1000)
res = model.predict(test_x)
res = tf.squeeze(res,1)
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=test_x, y=res,mode="markers"))
fig.show()  
```

#### 有一部分感觉拟合的不错，但总体没有很好的拟合，我觉得是sigmoid的范围在\[0,1\]的原因，去掉激活函数试试。

# Summary

intuitive:

1. 输出的激活函数非常重要，代表映射的值域。

2. 有一定拟合能力但误差较大时，可以考虑通过增加节点数或层数来提升性能

3. 不能拟合缺乏数据的地方，更像是泰勒拟合，没有学到周期性。
